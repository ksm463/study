```python
# 문제 12
# hp 변수에 대해 데이터표준화(Z-score)를 진행하고 이상치의 수를 구하시오.
# (단, 이상치는 z-score의 값이 1.5 초과하거나 -1.5 미만인 값이다.)
df = pd.read_csv("mtcars.csv")
```

```python
# 풀이
# Z = (X-평균) / 표준편차
std = df['hp'].std()
mean_hp = df['hp'].mean()
df['zscore'] = (df['hp']-mean_hp)  / std

cond1 = (df['zscore']>1.5)
cond2 = (df['zscore']<-1.5)
print(len(df[cond1])+len(df[cond2]))
```

```python
# 문제 13
# mpg 칼럼을 최소최대 스케일링을 진행한 후에 0.7보다 큰 값을 가지는 레코드 수를 구하라.
df = pd.read_csv("mtcars.csv")
```

```python
# 풀이
# 공식 : (x-min)/(max-min)
min_mpg = df['mpg'].min()
max_mpg = df['mpg'].max()
cond1 = (df['mpg']-min_mpg) / (max_mpg-min_mpg)
cond2 = cond1>0.7
print(len(df[cond2]))
```

```python
# 문제 14
# wt컬럼에 대해 상자그림 기준으로 이상치의 개수를 구하시오.
df = pd.read_csv("mtcars.csv")
```

```python
# (풀이)
# 이상치 : Q1, Q3 으로부터 1.5*IQR을 넘어가는 값
Q1 = df['wt'].quantile(0.25)
Q3 = df['wt'].quantile(0.75)
IQR = Q3-Q1

upper = Q3+(1.5*IQR)
lower = Q1-(1.5*IQR)

cond1 = (df['wt']>upper)
cond2 = (df['wt']<lower)

print(len(df[cond1])+len(df[cond2]))
```

```python
# 문제 19
# hp항목의 상위 7번째 값으로 상위 7개 값을 변환하고
# hp가 150 이상인 데이터를 추출하여 hp평균값을 반올림하여 정수로 출력하시오.
df = pd.read_csv('mtcars.csv')
```

```python
# (풀이)
# hp열 기준으로 내림차순 정렬
df = df.sort_values('hp', ascending=False)

# 인덱스 초기화
df = df.reset_index(drop=True) # drop=True는 기존 인덱스 삭제
# print(df)

# hp 상위 7번째 값 확인
top7 = df['hp'].loc[6] # top7 = 205

# np.where 활용
# np.where(조건, 조건에 해당할 때 값, 그렇지 않을 때 값)
import numpy as np
df['hp'] = np.where(df['hp']>=205, top7, df['hp'])

# hp 150이상인 데이터
cond1 = (df['hp']>=150)
df = df[cond1]
print(round(df['hp'].mean()))
```
